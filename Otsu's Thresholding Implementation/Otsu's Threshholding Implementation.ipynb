```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def plot_histogram(image,title):
  hist,_ = np.histogram(image.flatten(),256,[0,256])
  hist=hist/hist.sum()                # here i am normalizeing the pixel density
  plt.bar(np.arange(256), hist, color='orange')
  plt.xlabel('Intensity Value')
  plt.ylabel('Pixel density')
  plt.title(title)

def otsu_thresholding(image):
  hist, _ = np.histogram(image.flatten(), 256, [0, 256])
  hist = hist / hist.sum()
  prob = np.cumsum(hist)
  cum_mean = np.cumsum(hist * np.arange(256))
  global_mean = cum_mean[-1]
  weight1 = prob
  weight2 = 1 - prob
  mean1 = cum_mean / weight1
  mean2 = (global_mean - cum_mean) / weight2
  variance = weight1 * weight2 * (mean1 - mean2) ** 2    # this is the formula for In between Variance
  threshold = np.argmax(variance)                        # here we are maximising the in between variance
  binary_image = np.where(image > threshold, 255, 0)     # if the value of the pixel intensity is greater than threshhold, it will be mapped to 255 , else 0
  return binary_image.astype(np.uint8)                   # returning the binary image (dark pixel become completely black and light becomes complete white)

def add_gaussian_noise(image, variance):
    row, col = image.shape
    mean = 0
    gaussian = np.random.normal(mean, variance, (row, col))
    noisy_image = np.clip(image + gaussian, 0, 255)        # clip is that the intensity range is always between 0 and 255 and we are adding gaussian noise
    return noisy_image.astype(np.uint8)



```

```python
image_1=cv2.imread('/content/drive/MyDrive/tom.jpeg',0)
image_2=cv2.imread('/content/drive/MyDrive/panther.jpeg',0)
image_3=cv2.imread('/content/drive/MyDrive/bookpage_1.jpeg',0)
image_4=cv2.imread('/content/drive/MyDrive/bookpage_2.jpeg',0)
```

```python
binary_image=otsu_thresholding(image_1)
noisy_image=add_gaussian_noise(image_1,50)
noisy_binarised_image= otsu_thresholding(noisy_image)

plt.figure(figsize=(15,20))

plt.subplot(4,2,1)
plt.imshow(cv2.cvtColor(image_1, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.title('Given Original Image')

plt.subplot(4,2,2)
plot_histogram(image_1,'Histogram For Given Image')

plt.subplot(4,2,3)
plt.imshow(cv2.cvtColor(binary_image, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.title('Binarized Image')

plt.subplot(4,2,4)
plot_histogram(binary_image,'Histogram For Binarized Image')
plt.xticks([0,255])

plt.subplot(4,2,5)
plt.imshow(cv2.cvtColor(noisy_image,cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.title("Noisy image with Variance 50")

plt.subplot(4,2,6)
plot_histogram(noisy_image,'Histogram For Noisy Image with Var=50')

plt.subplot(4,2,7)
plt.imshow(cv2.cvtColor(noisy_binarised_image,cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.title("Binarised Noisy image")

plt.subplot(4,2,8)
plot_histogram(noisy_binarised_image,'Histogram For Binarised Noisy Image')
plt.xticks([0,255])

plt.show()
```
